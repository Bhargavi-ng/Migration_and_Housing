```python
import pandas as pd
```

```python
# Read in the excel
read_file = pd.read_excel (".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.xlsx")
# Write the dataframe object into csv file without the header
read_file.to_csv (".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv", index = None, header=False)
# Read csv file and convert into a dataframe object
cleaned_county_df = pd.DataFrame(pd.read_csv(".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv"))
# View the dataframe
cleaned_county_df.head()
```

```python
#Pick relevant columns into a new dataframe
county_df=cleaned_county_df[["State Name of Geography A","County Name of Geography A","State and County of Geography A",
                            "State/U.S. Island Area/Foreign Region of Geography B",
                            "County Name of Geography B","State and County of Geography B","Flow from Geography B to Geography A",
                            "Counterflow from Geography A to Geography B1","Net Migration from Geography B to Geography A1",
                            "Gross Migration between Geography A and Geography B1"]]
county_df.head()
```

```python
#Rename columns and replace spaces with underscore
county_df.rename(columns = {"State Name of Geography A":"State_Name_of_Geography_A",
                             "County Name of Geography A":"County_Name_of_Geography_A",
                             "State and County of Geography A":"State_and_County_of_Geography_A",
                            "State/U.S. Island Area/Foreign Region of Geography B":"State/U.S._Island_Area/Foreign_Region_of_Geography_B",
                            "County Name of Geography B":"County_Name_of_Geography_B",
                            "State and County of Geography B":"State_and_County_of_Geography_B",
                             "Flow from Geography B to Geography A":"Flow_from_Geography_B_to_Geography_A",
                            "Counterflow from Geography A to Geography B1":"Counterflow_from_Geography_A_to_Geography_B1",
                             "Net Migration from Geography B to Geography A1":"Net_Migration_from_Geography_B_to_Geography_A1",
                            "Gross Migration between Geography A and Geography B1":"Gross_Migration_between_Geography_A_and_Geography_B1"}
                           , inplace = True)
county_df.head()
```

```python
county_df.to_csv(".\Resources\cleaned-county-to-county-2015-2019-ins-outs-nets-gross.csv", header=True, index=False)
```
