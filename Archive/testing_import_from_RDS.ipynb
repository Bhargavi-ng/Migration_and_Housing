```python
# import dependencies
import pandas as pd
import psycopg2 as ps
from config import host_name,username, password, dbname
```

```python
# Create connection to RDS
def connect_to_db(hostname, db, port, user, pw):
   try:
       conn = ps.connect(host=hostname, database=db, user=user, password=pw, port=port)
 
   except ps.OperationalError as e:
       raise e
   else:
       print('Connected!')
       return conn
```

```python
# initialize connection
conn = None

conn = connect_to_db(host_name, dbname, 5432, username, password)
```

```python
curr = conn.cursor()
```

```python
# create list for dataframe column names
## importing from RDS brings in data, but doesn't pull in column names very easily
col_names = ['region_name','state_name','state_and_county','region_code']
```

```python
# appending additional column names

# for loop to append years 2000-2020 to column names list
for i in range(2000, 2021):
    col_names.append(str(i))

# appending percent change column name
col_names.append('percent_price_change_2015_19')

col_names
```

```python
# load data from RDS into dataframe
curr.execute("select * from house_price")
data=curr.fetchall()
df=pd.DataFrame(data, columns=col_names)
conn.close()
df.head()
```
