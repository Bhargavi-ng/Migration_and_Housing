```python
# Dropping first two rows and re-indexing
current_residence_df = current_residence_df.drop([0,1], axis=0, inplace=False)
current_residence_df = current_residence_df.reset_index(drop=True)
current_residence_df.head()
```

```python
# View list of columns after renaming
list(current_residence_df.columns.values)
```

```python
#dropping non-relevant columns into a new dataframe
cleaned_cr_df=current_residence_df.drop(columns=["Current_Residence_State_Code",
                                                "Current_Residence_FIPS_County_Code",
                                                "Residence_1_Year_Ago_State/U.S._Island Area/Foreign_Region_Code",
                                                "Residence_1_Year_Ago_FIPS_County_Code",
                                                "County_of_Current_Residence_Population_1_Year_and_Over_MOE",
                                                "County_of_Current_Residence_Nonmovers_MOE",
                                                "County_of_Current_Residence_Movers_within_United_States_1",
                                                "County_of_Current_Residence_Movers_within_Same_County_MOE",
                                                "County_of_Current_Residence_Movers_from_Different_County_Same_State_MOE",
                                                "County_of_Current_Residence_Movers_from_Different_State_MOE",
                                                "County_of_Current_Residence_Movers_from_Abroad_MOE",
                                                "County_of_Residence_1_Year_Ago_Population_1_Year_and_Over_MOE",
                                                "County_of_Residence_1_Year_Ago_Nonmovers_MOE",
                                                "County_of_Residence_1_Year_Ago_Movers_within_United_States_or_Puerto_Rico_MOE",
                                                "County_of_Residence_1_Year_Ago_Movers_within_Same_County_MOE",
                                                "County_of_Residence_1_Year_Ago_Movers_to_Puerto_Rico_MOE",
                                                "Movers_in_County_to_County_Flow_MOE"], inplace=False)
cleaned_cr_df.head()
```

```python
# View list of columns after dropping columns
list(cleaned_cr_df.columns.values)
```

```python
cleaned_cr_df.dtypes
```

```python
# Add a “County and state of current residence” column and “County and state of earlier residence” column
cleaned_cr_df['County_and_State_of_Current_Residence'] = cleaned_cr_df[['County_of_Current_Residence','State_of_Current_Residence']].apply(lambda x: ', '.join(x), axis=1)
cleaned_cr_df.head()
```

```python
# Add a “County and state of earlier residence” column
cleaned_cr_df['County_and_State_of_Previous_Residence'] = cleaned_cr_df[['County_of_Residence_1_Year_Ago', 'State/U.S._Island_Area/Foreign_Region_of_Residence_1_Year_Ago']].agg(','.join, axis=1)
cleaned_cr_df.head()
```

```python
# Export to csv file
cleaned_cr_df.to_csv(".\Resources\cleaned-county-to-county-2015-2019-current-residence-sort.csv", header=True, index=False)
```
