```python
import pandas as pd
```

```python
workbook_url = '.\Resources\county-to-county-2015-2019-current-residence-sort.xlsx'
```

```python
# List of States in the worksheet
states_list = ["Alabama", "Alaska", "Arizona", "Arkansas", "California", "Colorado", "Connecticut", "Delaware", 
               "District of Columbia", "Florida", "Georgia", "Hawaii", "Idaho", "Illinois", "Indiana", "Iowa", "Kansas",
               "Kentucky", "Louisiana", "Maine", "Maryland", "Massachusetts", "Michigan", "Minnesota", "Mississippi", 
               "Missouri", "Montana", "Nebraska", "Nevada", "New Hampshire", "New Jersey", "New Mexico", "New York", 
               "North Carolina", "North Dakota", "Ohio", "Oklahoma", "Oregon", "Pennsylvania", "Rhode Island", 
               "South Carolina", "South Dakota", "Tennessee", "Texas", "Utah", "Vermont", "Virginia", "Washington", 
               "West Virginia", "Wisconsin", "Wyoming", "Puerto Rico"
              ]
```

```python
#states_list = ["Alabama", "Alaska", "Arizona", "Puerto Rico"]
```

```python
# define global dataframe
clean_df = pd.DataFrame(columns = ['Current_Residence_State',
                                 'Current_Residence_County',
                                 'Current_Residence_County_Population',
                                 'Current_Residence_County_Nonmovers',
                                 'Current_Residence_Movers_within_US',
                                 'Current_Residence_Movers_within_Same_County',
                                 'Current_Residence_Movers_from_Different_County_Same_State',
                                 'Current_Residence_Movers_from_Different_State',
                                 'Current_Residence_Movers_from_Abroad',
                                 'Previous_Residence_Movers_within_US_or_PR',
                                 'Previous_Residence_County',
                                 'Previous_Residence_County_Population',
                                 'Previous_Residence_County_Nonmovers',
                                 'Previous_Residence_Movers_within_United_States_or_Puerto_Rico',
                                 'Previous_Residence_Movers_within_Same_County',
                                 'Previous_Residence_Movers_to_Different_County_within_Same_State',
                                 'Previous_Residence_Movers_to_Different_State',
                                 'Previous_Residence_Movers_to_PR',
                                 'Movers_in_County_to_County_Flow'])
```

```python
# loop through all worksheets in excel
for index, value in enumerate(states_list):
    # read data from one work sheet into the data frame.
    sheet_df = pd.read_excel(workbook_url, sheet_name = value)
    # rename the columns
    #sheet_df = sheet_df.apply(column_rename)
    sheet_df.rename(columns = {"Table with column headers in rows 2 through 4.":"Current_Residence_State_Code",
                            "Unnamed: 1":"Current_Residence_FIPS_County_Code",
                            "Unnamed: 2":"Previous_Residence_State/U.S._Island Area/Foreign_Region_Code",
                            "Unnamed: 3":"Previous_Residence_FIPS_County_Code",
                            "Unnamed: 4":"Current_Residence_State",
                            "Unnamed: 5":"Current_Residence_County",
                            "Unnamed: 6":"Current_Residence_County_Population",
                            "Unnamed: 7":"Current_Residence_County_Population_1_Year_and_Over_MOE",
                            "Unnamed: 8":"Current_Residence_County_Nonmovers",
                            "Unnamed: 9":"Current_Residence_County_Nonmovers_MOE",
                            "Unnamed: 10":"Current_Residence_Movers_within_US",
                            "Unnamed: 11":"Current_Residence_Movers_within_United_States_1",
                            "Unnamed: 12":"Current_Residence_Movers_within_Same_County",
                            "Unnamed: 13":"Current_Residence_Movers_within_Same_County_MOE",
                            "Unnamed: 14":"Current_Residence_Movers_from_Different_County_Same_State",
                            "Unnamed: 15":"Current_Residence_Movers_from_Different_County_Same_State_MOE",
                            "Unnamed: 16":"Current_Residence_Movers_from_Different_State",
                            "Unnamed: 17":"Current_Residence_Movers_from_Different_State_MOE",
                            "Unnamed: 18":"Current_Residence_Movers_from_Abroad",
                            "Unnamed: 19":"Current_Residence_Movers_from_Abroad_MOE",
                            "Unnamed: 20":"Previous_Residence_Movers_within_US_or_PR",
                            "Unnamed: 21":"Previous_Residence_County",
                            "Unnamed: 22":"Previous_Residence_County_Population",
                            "Unnamed: 23":"Previous_Residence_County_Population_1_Year_and_Over_MOE",
                            "Unnamed: 24":"Previous_Residence_County_Nonmovers",
                            "Unnamed: 25":"Previous_Residence_County_Nonmovers_MOE",
                            "Unnamed: 26":"Previous_Residence_Movers_within_United_States_or_Puerto_Rico",
                            "Unnamed: 27":"Previous_Residence_Movers_within_United_States_or_Puerto_Rico_MOE",
                            "Unnamed: 28":"Previous_Residence_Movers_within_Same_County",
                            "Unnamed: 29":"Previous_Residence_Movers_within_Same_County_MOE",
                            "Unnamed: 30":"Previous_Residence_Movers_to_Different_County_within_Same_State",
                            "Unnamed: 31":"Previous_Residence_Movers_within_Different_County_Same_State_MOE",
                            "Unnamed: 32":"Previous_Residence_Movers_to_Different_State",
                            "Unnamed: 33":"Previous_Residence_Movers_within_Different_State_MOE",
                            "Unnamed: 34":"Previous_Residence_Movers_to_PR",
                            "Unnamed: 35":"Previous_Residence_Movers_to_Puerto_Rico_MOE",
                            "Unnamed: 36":"Movers_in_County_to_County_Flow",
                            "Unnamed: 37":"Movers_in_County_to_County_Flow_MOE"}, inplace = True)
    
    # Drop first two rows and re-index
    sheet_df = sheet_df.drop([0,1,2], axis=0, inplace=False)
    sheet_df = sheet_df.reset_index(drop=True)
    # Drop non-relevant columns into a new dataframe
    sheet_df=sheet_df.drop(columns=["Current_Residence_State_Code",
                                                "Current_Residence_FIPS_County_Code",
                                                "Previous_Residence_State/U.S._Island Area/Foreign_Region_Code",
                                                "Previous_Residence_FIPS_County_Code",
                                                "Current_Residence_County_Population_1_Year_and_Over_MOE",
                                                "Current_Residence_County_Nonmovers_MOE",
                                                "Current_Residence_Movers_within_United_States_1",
                                                "Current_Residence_Movers_within_Same_County_MOE",
                                                "Current_Residence_Movers_from_Different_County_Same_State_MOE",
                                                "Current_Residence_Movers_from_Different_State_MOE",
                                                "Current_Residence_Movers_from_Abroad_MOE",
                                                "Previous_Residence_County_Population_1_Year_and_Over_MOE",
                                                "Previous_Residence_County_Nonmovers_MOE",
                                                "Previous_Residence_Movers_within_United_States_or_Puerto_Rico_MOE",
                                                "Previous_Residence_Movers_within_Same_County_MOE",
                                                "Previous_Residence_Movers_within_Different_County_Same_State_MOE",
                                                "Previous_Residence_Movers_within_Different_State_MOE",
                                                "Previous_Residence_Movers_to_Puerto_Rico_MOE",
                                                "Movers_in_County_to_County_Flow_MOE"], inplace=False)
    # Drop footers
    sheet_df = sheet_df.head(-7)
    # Fill NaNs with 0
    sheet_df = sheet_df.fillna(0)
    # Replace '-' with 0
    sheet_df = sheet_df.replace('-', 'NA')
    # Append to global dataframe
    clean_df = pd.concat([clean_df, sheet_df])

# View first 10 rows of cleaned file
clean_df.head(10)   
```

```python
# Add a “County and state of current residence” column 
clean_df['Current_Residence_County_State'] = clean_df['Current_Residence_County'] +', '+ clean_df['Current_Residence_State']
clean_df.head()
```

```python
# Add a “County and state of Previous residence” column
clean_df['Previous_Residence_County_State'] = clean_df['Previous_Residence_County'] +', '+ clean_df['Previous_Residence_Movers_within_US_or_PR']
clean_df.head() 
```
